```python
# Define a function to remove outliers based on the IQR method
def remove_outliers(df, columns):
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1
    
    # Filtering the data that falls within the IQR
    df_cleaned = df[~((df[columns] < (Q1 - 1.5 * IQR)) | (df[columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df_cleaned

# List of numerical columns to check for outliers
numeric_columns = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']

# Remove outliers from the DataFrame
df_no_outliers = remove_outliers(df, numeric_columns)

# Display the shape of the DataFrame before and after removing outliers
df.shape, df_no_outliers.shape
```

```python
# Save the cleaned DataFrame (after removing outliers) to a new CSV file
output_path = 'cleaned_data.csv'
df_no_outliers.to_csv(output_path, index=False)

print(f"Cleaned dataset saved as {output_path}")
```

```python
# Convert the 'type' column to numerical form using label encoding
df['type_numeric'] = df['type'].astype('category').cat.codes

# Display the first few rows to verify the transformation
df[['type', 'type_numeric']].head()
```

```python
from sklearn.model_selection import train_test_split

# Defining the selected features (excluding identifiers like nameOrig, nameDest)
features_selected = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest', 'type_numeric']

# Defining the feature matrix (X) and the target variable (y)
X_selected = df[features_selected]
y_selected = df['isFraud']

# Splitting the data into training and testing sets (80% train, 20% test)
X_train_selected, X_test_selected, y_train_selected, y_test_selected = train_test_split(X_selected, y_selected, test_size=0.2, random_state=42)

# Display the shapes of the train and test sets to ensure correct split
print("Training Features Shape:", X_train_selected.shape)
print("Test Features Shape:", X_test_selected.shape)
print("Training Target Shape:", y_train_selected.shape)
print("Test Target Shape:", y_test_selected.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
model.fit(X_train_selected, y_train_selected)

# Make predictions on the test data
y_pred_selected = model.predict(X_test_selected)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test_selected, y_pred_selected)
classification_rep = classification_report(y_test_selected, y_pred_selected)
confusion_mat = confusion_matrix(y_test_selected, y_pred_selected)

# Print the evaluation results
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:\n", classification_rep)
print("Confusion Matrix:\n", confusion_mat)
```
